```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.
# Choices made: seed set to 1847 (your Drexel ID last 4 digits) for reproducibility;
# timestamps are sorted so plots progress in time; markers+lines chosen for clarity.
import numpy as np
import matplotlib.pyplot as plt
# Seed: last 4 digits of Drexel ID provided by user
SEED = 1847
rng = np.random.default_rng(SEED)
n = 200
# generate timestamps uniformly on [0, 10]
t = rng.uniform(0, 10, size=n)
# generate sensor readings from gaussian distributions
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n)
# sort by time so plotting shows evolution over time
order = np.argsort(t)
t = t[order]
sensor_a = sensor_a[order]
sensor_b = sensor_b[order]
# plotting
plt.figure(figsize=(8, 4))
plt.scatter(t, sensor_a, label="Sensor A", s=20, alpha=0.7, edgecolors="none")
plt.scatter(t, sensor_b, label="Sensor B", s=20, alpha=0.7, marker="s", edgecolors="none")
plt.xlabel("Time (s)")
plt.ylabel("Temperature (°C)")
plt.title(f"Simulated Temperature Readings (seed={SEED})")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: sensor reading (y-axis) vs timestamp in seconds (x-axis).
# Color Sensor A points blue and Sensor B points orange.
# Add axis labels with units, a title, and a legend.
```

```python
# Overlaid histogram of Sensor A and Sensor B temperature distributions.
plt.figure(figsize=(8, 4))
bins = 30
plt.hist(sensor_a, bins=bins, alpha=0.5, color='blue', label='Sensor A')
plt.hist(sensor_b, bins=bins, alpha=0.5, color='orange', label='Sensor B')
# vertical lines at each sensor's mean
mean_a = np.mean(sensor_a)
mean_b = np.mean(sensor_b)
plt.axvline(mean_a, color='blue', linestyle='--', linewidth=1, label=f'Sensor A mean = {mean_a:.2f} °C')
plt.axvline(mean_b, color='orange', linestyle='--', linewidth=1, label=f'Sensor B mean = {mean_b:.2f} °C')
plt.xlabel('Temperature (°C)')
plt.ylabel('Count')
plt.title(f'Histogram of Sensor Readings (seed={SEED})')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Side-by-side box plot comparing Sensor A and Sensor B distributions.
plt.figure(figsize=(6, 6))
data = [sensor_a, sensor_b]
bp = plt.boxplot(data, labels=["Sensor A", "Sensor B"], patch_artist=True)
colors = ["lightblue", "lightgreen"]
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
# compute overall mean across both sensors and draw a horizontal dashed line
overall_mean = np.mean(np.concatenate(data))
plt.axhline(overall_mean, color='gray', linestyle='--', linewidth=1, label=f'Overall mean = {overall_mean:.2f} °C')
plt.xlabel("Sensor")
plt.ylabel("Temperature (deg C)")
plt.title(f"Side-by-side Box Plot of Sensor Readings (seed={SEED})")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
